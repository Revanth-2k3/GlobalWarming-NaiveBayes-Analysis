```python
# Drop rows with missing values in critical columns
cleaned_data = data.dropna(subset=["Temperature_Anomaly", "CO2_Emissions", "Population"])
```

```python
# Create a target variable by classifying Temperature_Anomaly as 'High' or 'Low'
threshold = 0  # Example threshold for categorization
cleaned_data["Anomaly_Class"] = cleaned_data["Temperature_Anomaly"].apply(
    lambda x: "High" if x > threshold else "Low"
)
```

```python
# Define features and target
features = cleaned_data[["CO2_Emissions", "Population", "Renewable_Energy_Usage"]]
target = cleaned_data["Anomaly_Class"]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

# Create and train the Naïve Bayes classifier
nb_classifier = GaussianNB()
nb_classifier.fit(X_train_scaled, y_train)
```

```python
# Predict on the test set
y_pred = nb_classifier.predict(X_test_scaled)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
# Print results
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:\n")
print(report)
```
